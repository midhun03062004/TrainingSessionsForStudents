```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data (replace with your actual data)
# Example: Predicting house price based on square footage and number of bedrooms
X = np.array([[1000, 2], [1500, 3], [1200, 2], [1800, 4], [2000, 3]])  # Features (square footage, bedrooms)
y = np.array([250000, 350000, 300000, 400000, 420000])  # Target (house prices)

# Create a Linear Regression model
model = LinearRegression()

# Train the model (fit the data)
model.fit(X, y)

# Get the coefficients (slopes) and intercept
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)

# Make predictions
new_data = np.array([[1600, 3], [2200, 4]])  # Example new data
predictions = model.predict(new_data)

print("Predictions:", predictions)

#Example of accessing single values.
print(f"Coefficient for square footage: {coefficients[0]}")
print(f"Coefficient for number of bedrooms: {coefficients[1]}")
```
